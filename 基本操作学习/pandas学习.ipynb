# Missing Data
np.nan代表缺失的数据

reindex()来重新构建df
* To drop any rows that have missing data.--->`dropna()`
* Filling missing data.--->`pd.isna()`,`fillna()`

```python
k2 = k.reindex(index=dates[0:4], columns=list(k.columns) + ['G'])
```

```python
k2
```

```python
k2.dropna(how='any')        #行删除
```

```python
k2.fillna(value='None')        #填充缺失值
```

```python
pd.isna(k2)         #用布尔值查看缺失值
```

# Reshaping

```python
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz',
   ....:                      'foo', 'foo', 'qux', 'qux'],
   ....:                     ['one', 'two', 'one', 'two',
   ....:                      'one', 'two', 'one', 'two']]))
```

```python
tuples
```

```python
index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
```

```python
index
```

```python
df
```

```python
df2 = df[:4]
```

```python
df2
```

```python
stacked = df2.stack()               #宽变长
```

# Categoricals
* 将数据转为分类标签df.astype('')
>把某个属性当作分类变量：
>* 用`astype('catagory')`
>* 然后将该分类标签重新命名`cat.categories`
>* 替换原来的列

```python
df = pd.DataFrame({"id":[1,2,3,4,5,6], "raw_grade":['a', 'b', 'b', 'a', 'a', 'e']})
```

```python
df
```

```python
df['grade'] = df["raw_grade"].astype("category")           #astype
```

```python
df['grade']
```

```python
df["grade"].cat.categories = ["very good", "good", "very bad"]
```

```python
df
```

```python
df['grade']
```

```python
df.sort_values(by='grade',ascending=False)
```

```python
df.groupby('grade').size()
```
